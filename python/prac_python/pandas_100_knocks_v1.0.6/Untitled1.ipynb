```python
from pandas_datareader import data
import pandas as pd
```

```python
stock_list = {"MAXIS-FTSE": 2559, "MAXIS-SP500": 2558, "MAXIS-NASDAQ100": 2631, "MAXIS-TOPIX": 1348, "新興国": 1681, "東証REIT": 1343, "世界REIT": 2515, "LMBA-GOLD": "GOLD11.SA" }
concat_list = []
```

```python
for name, code in stock_list.items():
    try:
        int(code)
        df_raw = data.DataReader(f"{code}.T","yahoo", start="2020-07-01", end="2022-07-01")
    except:
        df_raw = data.DataReader(f"{code}","yahoo", start="2020-07-01", end="2022-07-01")
    df_close = df_raw["Close"]
    concat_list.append(df_close)
```

```python
df_total = pd.concat(concat_list, axis=1)
df_total.columns = stock_list.keys()
```

```python
df_corr = df_total.corr()
df_corr
```
