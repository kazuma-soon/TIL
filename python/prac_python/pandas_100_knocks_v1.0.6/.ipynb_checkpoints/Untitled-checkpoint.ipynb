```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
def numerical_gradient(f, x):
    h = 1e-4
    grad = np.zeros_like(x)
    
    for idx in range(x.size):
        tmp_x = x[idx]
        # f(x + h)
        x[idx] = tmp_x + h
        fxh1 = f(x)
        
        # f(x - h)
        x[idx] = tmp_x - h
        fxh2 = f(x)
        
        grad[idx] = (fxh1 - fxh2) / (2*h)
        x[idx] = tmp_x
    
    return grad
```

```python
def function2(x):
    return x[0]**2 + x[1]**2
```

```python
def gradient_discent(f, init_x, lr=0.01, step_num=100):
    x = init_x
    
    for i in range(step_num):
        grad = numerical_gradient(f, x)
        x -= (lr * grad)
        
    return x
```

```python
init_x = np.array([-389, 39999999.0])
gradient_discent(function2, init_x=init_x, lr=0.1, step_num=100)
```
